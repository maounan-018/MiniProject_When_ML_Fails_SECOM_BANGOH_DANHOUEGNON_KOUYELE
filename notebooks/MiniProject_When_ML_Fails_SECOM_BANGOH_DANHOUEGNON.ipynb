# Mini-Project — When Machine Learning Fails
## Diagnosing and Repairing Failure Modes of a Non-Linear Model

**Dataset :** SECOM  
**Failure mode principal :** shortcut learning / corrélation non causale  
**Modes d'echec bonus :** class imbalance failure, distribution shift entre phases industrielles  
**Modèle non linéaire :** Random Forest  
**Auteurs :** David Danhouegnon & Levis BANGOH


## 1. Research question and chosen dataset

Nous utilisons le dataset **SECOM**, autorisé par le sujet. La cible `Pass/Fail` vaut `-1` pour `Pass` et `1` pour `Fail`.

**Question de recherche :**

> Dans quelle mesure la dépendance d'un Random Forest à un raccourci non causal augmente-t-elle avec la fiabilité apparente de ce raccourci dans l'environnement d'entraînement ? Cette dépendance provoque-t-elle une chute mesurable lorsque la corrélation est cassée en test ?

La question est falsifiable : si l'importance du shortcut reste faible, ou si la performance reste stable quand il est randomisé/inversé, l'hypothèse est rejetée.


```python
# ============================================================
# Mini-Project — When Machine Learning Fails
# Dataset SECOM — Shortcut learning sur modèle non linéaire
# ============================================================

import warnings
warnings.filterwarnings('ignore')

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, balanced_accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix, classification_report,
    precision_recall_curve,
)
from sklearn.inspection import permutation_importance

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_context('notebook')
```

```python
# --- Chargement et exploration ---
# Le CSV est inclus dans le depot GitHub, dans le dossier data/.

CSV_NAME = 'uci-secom data set ML.csv'
cwd = Path.cwd()
DATA_PATH_CANDIDATES = [
    cwd / 'data' / CSV_NAME,
    cwd.parent / 'data' / CSV_NAME,
    cwd / CSV_NAME,
    cwd.parent / CSV_NAME,
]

DATA_PATH = next((path for path in DATA_PATH_CANDIDATES if path.exists()), None)
if DATA_PATH is None:
    checked = '\n'.join(str(path) for path in DATA_PATH_CANDIDATES)
    raise FileNotFoundError(f'Dataset SECOM introuvable. Chemins testes :\n{checked}')

df_raw = pd.read_csv(DATA_PATH)
print(f'Dataset utilise : {DATA_PATH}')
print(f'Dimensions : {df_raw.shape}')
print('\nDistribution de la cible :')
print(df_raw['Pass/Fail'].value_counts())
print('\nDistribution de Phase :')
print(df_raw['Phase'].value_counts(dropna=False))
display(df_raw.head())
```

```python
# --- Préparation sans leakage ---
def prepare_secom_features(df, missing_threshold=0.50):
    df = df.copy()
    y = (df['Pass/Fail'] == 1).astype(int)
    X = df.drop(columns=['Pass/Fail'])
    if 'Time' in X.columns:
        X = X.drop(columns=['Time'])
    high_missing = X.isna().mean()[lambda s: s > missing_threshold].index.tolist()
    X = X.drop(columns=high_missing)
    X = pd.get_dummies(X, columns=['Phase'], dummy_na=False)
    constant_cols = X.nunique(dropna=True)[lambda s: s <= 1].index.tolist()
    X = X.drop(columns=constant_cols)
    return X, y, {'high_missing': high_missing, 'constant_cols': constant_cols}

X_clean, y, prep_report = prepare_secom_features(df_raw)
X_train_clean, X_test_clean, y_train, y_test = train_test_split(
    X_clean, y, test_size=0.30, random_state=RANDOM_SEED, stratify=y,
)
print(f"X_clean : {X_clean.shape}")
print(f"Train : {X_train_clean.shape}, Fail = {y_train.sum()} ({y_train.mean():.2%})")
print(f"Test  : {X_test_clean.shape}, Fail = {y_test.sum()} ({y_test.mean():.2%})")

def build_random_forest(seed=RANDOM_SEED, class_weight='balanced_subsample'):
    return Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('model', RandomForestClassifier(
            n_estimators=400, min_samples_leaf=2, class_weight=class_weight,
            random_state=seed, n_jobs=-1,
        )),
    ])

def evaluate_model(model, X_eval, y_eval, label):
    y_pred = model.predict(X_eval)
    y_proba = model.predict_proba(X_eval)[:, 1]
    return {
        'setting': label,
        'accuracy': accuracy_score(y_eval, y_pred),
        'balanced_accuracy': balanced_accuracy_score(y_eval, y_pred),
        'precision_fail': precision_score(y_eval, y_pred, zero_division=0),
        'recall_fail': recall_score(y_eval, y_pred, zero_division=0),
        'f1_fail': f1_score(y_eval, y_pred, zero_division=0),
        'roc_auc': roc_auc_score(y_eval, y_proba),
    }



def display_confusion_from_predictions(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred, labels=[0, 1])
    fig, ax = plt.subplots(figsize=(4.8, 4))
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        xticklabels=['Pass', 'Fail'],
        yticklabels=['Pass', 'Fail'],
        ax=ax,
    )
    ax.set_title(title)
    ax.set_xlabel('Prédiction')
    ax.set_ylabel('Vérité')
    plt.tight_layout()
    plt.show()


def display_confusion(model, X_eval, y_eval, title):
    display_confusion_from_predictions(y_eval, model.predict(X_eval), title)
```

## 2. Reference model and observed symptom

On entraîne d'abord un Random Forest sans raccourci non causal. Ensuite, on injecte `shortcut_alarm`, une variable artificielle fortement corrélée à la cible dans le train, puis on casse cette corrélation en test.


```python
# --- Modèle de référence ---
reference_model = build_random_forest()
reference_model.fit(X_train_clean, y_train)
reference_results = pd.DataFrame([
    evaluate_model(reference_model, X_train_clean, y_train, 'Reference — train clean'),
    evaluate_model(reference_model, X_test_clean, y_test, 'Reference — test clean'),
])
display(reference_results)
print(classification_report(y_test, reference_model.predict(X_test_clean), target_names=['Pass','Fail'], digits=4))
```

```python
# --- Création du shortcut contrôlé ---
def make_shortcut_feature(y_values, reliability, seed, mode='aligned'):
    rng = np.random.default_rng(seed)
    y_values = np.asarray(y_values).astype(int)
    if mode == 'aligned':
        shortcut = y_values.copy()
    elif mode == 'inverted':
        shortcut = 1 - y_values.copy()
    elif mode == 'random':
        return rng.integers(0, 2, len(y_values))
    else:
        raise ValueError('mode invalide')
    flip = rng.random(len(y_values)) > reliability
    shortcut[flip] = 1 - shortcut[flip]
    return shortcut

def add_shortcut(X, y_values, reliability, seed, mode):
    X_new = X.copy()
    X_new['shortcut_alarm'] = make_shortcut_feature(y_values, reliability, seed, mode)
    return X_new

SHORTCUT_RELIABILITY = 0.98
X_train_shortcut = add_shortcut(X_train_clean, y_train, SHORTCUT_RELIABILITY, RANDOM_SEED+1, 'aligned')
X_test_shortcut_id = add_shortcut(X_test_clean, y_test, SHORTCUT_RELIABILITY, RANDOM_SEED+2, 'aligned')
X_test_shortcut_random = add_shortcut(X_test_clean, y_test, SHORTCUT_RELIABILITY, RANDOM_SEED+3, 'random')
X_test_shortcut_ood = add_shortcut(X_test_clean, y_test, SHORTCUT_RELIABILITY, RANDOM_SEED+4, 'inverted')

shortcut_diagnostic = pd.DataFrame({
    'dataset': ['train aligned', 'test aligned', 'test random', 'test inverted'],
    'corr_with_target': [
        np.corrcoef(X_train_shortcut['shortcut_alarm'], y_train)[0,1],
        np.corrcoef(X_test_shortcut_id['shortcut_alarm'], y_test)[0,1],
        np.corrcoef(X_test_shortcut_random['shortcut_alarm'], y_test)[0,1],
        np.corrcoef(X_test_shortcut_ood['shortcut_alarm'], y_test)[0,1],
    ],
})
display(shortcut_diagnostic)

broken_model = build_random_forest()
broken_model.fit(X_train_shortcut, y_train)
broken_results = pd.DataFrame([
    evaluate_model(broken_model, X_train_shortcut, y_train, 'Broken — train shortcut aligned'),
    evaluate_model(broken_model, X_test_shortcut_id, y_test, 'Broken — test shortcut aligned'),
    evaluate_model(broken_model, X_test_shortcut_random, y_test, 'Broken — test shortcut random'),
    evaluate_model(broken_model, X_test_shortcut_ood, y_test, 'Broken — test shortcut inverted'),
])
display(broken_results)


# Symptôme explicite : matrices de confusion.
# On compare le même modèle cassé lorsque le shortcut est valide vs lorsqu'il est inversé.
display_confusion(broken_model, X_test_shortcut_id, y_test, 'Broken model — test shortcut aligned')
display_confusion(broken_model, X_test_shortcut_ood, y_test, 'Broken model — test shortcut inverted')

print('Rapport de classification — shortcut aligné')
print(classification_report(y_test, broken_model.predict(X_test_shortcut_id), target_names=['Pass', 'Fail'], digits=4))
print('\nRapport de classification — shortcut inversé')
print(classification_report(y_test, broken_model.predict(X_test_shortcut_ood), target_names=['Pass', 'Fail'], digits=4))
```

## 3. Causal hypothesis and controlled experiment

**Hypothèse causale :** le modèle dépend excessivement de `shortcut_alarm`.  
**Contrôle :** modèle de référence sans shortcut.  
**Test contrôlé :** même modèle, mêmes lignes, mais shortcut aligné, randomisé, puis inversé.


```python
# --- Sensibilité à la fiabilité du shortcut ---
reliability_levels = [0.55, 0.70, 0.85, 0.98]
rows = []
for i, reliability in enumerate(reliability_levels):
    Xtr = add_shortcut(X_train_clean, y_train, reliability, RANDOM_SEED+100+i, 'aligned')
    Xid = add_shortcut(X_test_clean, y_test, reliability, RANDOM_SEED+200+i, 'aligned')
    Xood = add_shortcut(X_test_clean, y_test, reliability, RANDOM_SEED+300+i, 'inverted')
    model = build_random_forest(seed=RANDOM_SEED+i)
    model.fit(Xtr, y_train)
    for context, Xev in [('ID aligned', Xid), ('OOD inverted', Xood)]:
        row = evaluate_model(model, Xev, y_test, f'reliability={reliability:.2f} — {context}')
        row['shortcut_reliability'] = reliability
        row['evaluation_context'] = context
        rows.append(row)
shortcut_sensitivity = pd.DataFrame(rows)
display(shortcut_sensitivity[['shortcut_reliability','evaluation_context','balanced_accuracy','recall_fail','f1_fail','roc_auc']])

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.lineplot(data=shortcut_sensitivity, x='shortcut_reliability', y='roc_auc', hue='evaluation_context', marker='o', ax=axes[0])
sns.lineplot(data=shortcut_sensitivity, x='shortcut_reliability', y='f1_fail', hue='evaluation_context', marker='o', ax=axes[1])
axes[0].set_ylim(0, 1.05)
axes[1].set_ylim(0, 1.05)
plt.tight_layout()
plt.show()
```

```python
# --- Importance native + permutation importance ---
rf = broken_model.named_steps['model']
native_importances = pd.DataFrame({
    'feature': X_train_shortcut.columns,
    'native_importance': rf.feature_importances_,
}).sort_values('native_importance', ascending=False)
display(native_importances.head(15))

perm = permutation_importance(
    broken_model, X_test_shortcut_id, y_test,
    scoring='roc_auc', n_repeats=10, random_state=RANDOM_SEED, n_jobs=-1,
)
perm_importances = pd.DataFrame({
    'feature': X_test_shortcut_id.columns,
    'permutation_importance_mean': perm.importances_mean,
    'permutation_importance_std': perm.importances_std,
}).sort_values('permutation_importance_mean', ascending=False)
display(perm_importances.head(15))

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
sns.barplot(data=native_importances.head(12), y='feature', x='native_importance', color='#4f81bd', ax=axes[0])
axes[0].set_title('Importances des fonctionnalités — modèle cassé')
sns.barplot(data=perm_importances.head(12), y='feature', x='permutation_importance_mean', color='#70ad47', ax=axes[1])
axes[1].set_title('Permutation importance — modèle cassé')
plt.tight_layout()
plt.show()
```

## 4. Proposed correction and evaluation

Correction ciblée : retirer explicitement `shortcut_alarm` du pipeline pollué.

Cette correction attaque la **cause** et non le symptôme : on ne change pas seulement le seuil de décision, ni le `class_weight`, ni une métrique d'évaluation. On supprime la variable non causale elle-même, c'est-à-dire la source de la dépendance au raccourci. Si l'hypothèse causale est correcte, le modèle corrigé doit perdre l'avantage artificiel en test in-distribution, mais retrouver une robustesse raisonnable lorsque la corrélation du shortcut est inversée.

Le modèle corrigé redevient proche du modèle de référence ; c'est attendu, car on retire précisément la cause identifiée.


```python
# --- Correction : retrait du shortcut ---
X_train_sanitized = X_train_shortcut.drop(columns=['shortcut_alarm'])
X_test_sanitized = X_test_shortcut_ood.drop(columns=['shortcut_alarm'])
corrected_model = build_random_forest()
corrected_model.fit(X_train_sanitized, y_train)

summary_before_after = pd.DataFrame([
    evaluate_model(broken_model, X_test_shortcut_id, y_test, 'Broken — ID shortcut aligned'),
    evaluate_model(broken_model, X_test_shortcut_ood, y_test, 'Broken — OOD shortcut inverted'),
    evaluate_model(corrected_model, X_test_sanitized, y_test, 'Corrected — shortcut removed'),
    evaluate_model(reference_model, X_test_clean, y_test, 'Reference — test clean'),
])
display(summary_before_after)

fig, ax = plt.subplots(figsize=(9, 5))
plot_df = summary_before_after.melt(
    id_vars='setting',
    value_vars=['balanced_accuracy','recall_fail','f1_fail','roc_auc'],
    var_name='metric', value_name='score'
)
sns.barplot(data=plot_df, x='metric', y='score', hue='setting', ax=ax)
ax.set_ylim(0, 1.05)
ax.set_title('Comparaison avant / après correction')
plt.tight_layout()
plt.show()
```

## 5. Experimental validation : variance entre seeds

Un seul split peut donner une conclusion fragile, surtout avec peu d'observations `Fail`. On répète donc l'expérience sur plusieurs seeds.


```python
def run_one_seed(seed, reliability=SHORTCUT_RELIABILITY):
    Xtr, Xte, ytr, yte = train_test_split(X_clean, y, test_size=0.30, random_state=seed, stratify=y)
    Xtr_sc = add_shortcut(Xtr, ytr, reliability, seed+1, 'aligned')
    Xte_id = add_shortcut(Xte, yte, reliability, seed+2, 'aligned')
    Xte_ood = add_shortcut(Xte, yte, reliability, seed+3, 'inverted')
    broken = build_random_forest(seed)
    broken.fit(Xtr_sc, ytr)
    corrected = build_random_forest(seed)
    corrected.fit(Xtr, ytr)
    out = pd.DataFrame([
        evaluate_model(broken, Xte_id, yte, 'broken_id'),
        evaluate_model(broken, Xte_ood, yte, 'broken_ood'),
        evaluate_model(corrected, Xte, yte, 'corrected_clean'),
    ])
    out['seed'] = seed
    return out

seed_results = pd.concat([run_one_seed(s) for s in [7, 11, 19, 23, 42]], ignore_index=True)
display(seed_results)
seed_summary = seed_results.groupby('setting').agg(
    accuracy_mean=('accuracy', 'mean'),
    accuracy_std=('accuracy', 'std'),
    balanced_accuracy_mean=('balanced_accuracy', 'mean'),
    balanced_accuracy_std=('balanced_accuracy', 'std'),
    recall_fail_mean=('recall_fail', 'mean'),
    recall_fail_std=('recall_fail', 'std'),
    f1_fail_mean=('f1_fail', 'mean'),
    f1_fail_std=('f1_fail', 'std'),
    roc_auc_mean=('roc_auc', 'mean'),
    roc_auc_std=('roc_auc', 'std'),
).round(4).reset_index()

display(seed_summary)
```

## 6. Failure mode A — Class imbalance failure

SECOM est naturellement déséquilibré. Un modèle non pondéré peut afficher une accuracy élevée mais mal détecter la classe `Fail`. Correction : `class_weight` + optimisation du seuil sur validation.


```python
X_train_inner, X_val_inner, y_train_inner, y_val_inner = train_test_split(
    X_train_clean, y_train, test_size=0.25, random_state=RANDOM_SEED, stratify=y_train,
)

imbalance_broken_model = build_random_forest(class_weight=None)
imbalance_broken_model.fit(X_train_inner, y_train_inner)
imbalance_broken_results = pd.DataFrame([
    evaluate_model(imbalance_broken_model, X_val_inner, y_val_inner, 'Imbalance broken — validation'),
    evaluate_model(imbalance_broken_model, X_test_clean, y_test, 'Imbalance broken — test'),
])

imbalance_corrected_model = build_random_forest()
imbalance_corrected_model.fit(X_train_inner, y_train_inner)
y_val_proba = imbalance_corrected_model.predict_proba(X_val_inner)[:, 1]
precisions, recalls, thresholds = precision_recall_curve(y_val_inner, y_val_proba)
f1_scores = 2 * precisions * recalls / (precisions + recalls + 1e-12)
best_threshold = thresholds[int(np.nanargmax(f1_scores[:-1]))]
print(f'Seuil optimisé sur validation : {best_threshold:.4f}')

def evaluate_with_threshold(model, X_eval, y_eval, threshold, label):
    y_proba = model.predict_proba(X_eval)[:, 1]
    y_pred = (y_proba >= threshold).astype(int)
    return {
        'setting': label, 'threshold': threshold,
        'accuracy': accuracy_score(y_eval, y_pred),
        'balanced_accuracy': balanced_accuracy_score(y_eval, y_pred),
        'precision_fail': precision_score(y_eval, y_pred, zero_division=0),
        'recall_fail': recall_score(y_eval, y_pred, zero_division=0),
        'f1_fail': f1_score(y_eval, y_pred, zero_division=0),
        'roc_auc': roc_auc_score(y_eval, y_proba),
    }

imbalance_corrected_results = pd.DataFrame([
    evaluate_with_threshold(imbalance_corrected_model, X_test_clean, y_test, 0.5, 'Corrected weighted — test threshold 0.5'),
    evaluate_with_threshold(imbalance_corrected_model, X_test_clean, y_test, best_threshold, 'Corrected weighted — test optimized'),
])
display(pd.concat([imbalance_broken_results, imbalance_corrected_results], ignore_index=True))
```

```python
# Matrices de confusion — bonus imbalance
# Symptôme : le modèle non pondéré écrase la classe Fail (recall ≈ 0)
# Correction : class_weight=balanced + seuil optimisé sur validation

y_pred_broken_imb = imbalance_broken_model.predict(X_test_clean)
y_proba_corrected_imb = imbalance_corrected_model.predict_proba(X_test_clean)[:, 1]
y_pred_corrected_imb = (y_proba_corrected_imb >= best_threshold).astype(int)

display_confusion_from_predictions(
    y_test, y_pred_broken_imb,
    'Imbalance broken — test (seuil 0.5, sans class_weight)'
)
display_confusion_from_predictions(
    y_test, y_pred_corrected_imb,
    f'Imbalance corrected — test (seuil optimisé {best_threshold:.3f}, class_weight=balanced)'
)

print('Rapport — modèle cassé (imbalance)')
print(classification_report(y_test, y_pred_broken_imb, target_names=['Pass', 'Fail'], digits=4))
print('Rapport — modèle corrigé (imbalance)')
print(classification_report(y_test, y_pred_corrected_imb, target_names=['Pass', 'Fail'], digits=4))
```

## 7. Failure mode B — Distribution shift entre phases industrielles

On tient hors entraînement une phase industrielle contenant assez d'exemples des deux classes. Le garde-fou évite les splits non stratifiables.


```python
phase_fail_counts = pd.crosstab(df_raw['Phase'], df_raw['Pass/Fail'])
display(phase_fail_counts)
eligible_phases = phase_fail_counts[(phase_fail_counts[-1] >= 10) & (phase_fail_counts[1] >= 4)]
if eligible_phases.empty:
    raise ValueError("Aucune phase ne contient assez d'exemples des deux classes pour un split OOD robuste.")

holdout_phase = eligible_phases[1].idxmax()
print(f'Phase tenue hors entraînement : {holdout_phase}')
phase_series = df_raw.loc[X_clean.index, 'Phase']
source_mask = phase_series != holdout_phase
target_mask = phase_series == holdout_phase
X_source, y_source = X_clean.loc[source_mask], y.loc[source_mask]
X_target, y_target = X_clean.loc[target_mask], y.loc[target_mask]

def safe_stratified_split(X_part, y_part, test_size, seed, label):
    counts = y_part.value_counts()
    if len(counts) < 2 or counts.min() < 2:
        raise ValueError(f'Split impossible pour {label}: classes insuffisantes {counts.to_dict()}')
    return train_test_split(X_part, y_part, test_size=test_size, random_state=seed, stratify=y_part)

X_source_train, X_source_test, y_source_train, y_source_test = safe_stratified_split(X_source, y_source, 0.30, RANDOM_SEED, 'source')
X_target_cal, X_target_test, y_target_cal, y_target_test = safe_stratified_split(X_target, y_target, 0.70, RANDOM_SEED, 'target')

shift_broken_model = build_random_forest()
shift_broken_model.fit(X_source_train, y_source_train)
shift_broken_results = pd.DataFrame([
    evaluate_model(shift_broken_model, X_source_test, y_source_test, 'Shift broken — source test'),
    evaluate_model(shift_broken_model, X_target_test, y_target_test, 'Shift broken — target phase test'),
])

X_shift_corrected_train = pd.concat([X_source_train, X_target_cal])
y_shift_corrected_train = pd.concat([y_source_train, y_target_cal])
shift_corrected_model = build_random_forest()
shift_corrected_model.fit(X_shift_corrected_train, y_shift_corrected_train)
shift_corrected_results = pd.DataFrame([
    evaluate_model(shift_corrected_model, X_source_test, y_source_test, 'Shift corrected — source test'),
    evaluate_model(shift_corrected_model, X_target_test, y_target_test, 'Shift corrected — target phase test'),
])

shift_summary = pd.concat([
    shift_broken_results.assign(model='Broken'),
    shift_corrected_results.assign(model='Corrected'),
], ignore_index=True)
display(shift_summary[['model','setting','accuracy','balanced_accuracy','precision_fail','recall_fail','f1_fail','roc_auc']])
```

## 8. Threats to validity

1. **Shortcut artificiel.** Il prouve un mécanisme causal contrôlé, pas une découverte directe dans SECOM.  
2. **Correction proche du modèle de référence.** C'est attendu, car la correction retire la seule feature non causale ajoutée.  
3. **Classe Fail rare.** Recall et F1 peuvent varier fortement selon le split.  
4. **Randomness.** Plusieurs seeds réduisent le risque, sans constituer une preuve statistique complète.  
5. **Correction partielle.** Retirer `shortcut_alarm` ne résout pas tous les défis : valeurs manquantes, haute dimensionnalité et déséquilibre.  
6. **Data leakage.** L'imputation est dans le pipeline, et les tests shortcut sont séparés du train.


## 9. Conclusion

Le Random Forest peut obtenir une excellente performance pour une mauvaise raison : il exploite `shortcut_alarm` lorsque cette variable est fortement corrélée à la cible. Quand la corrélation est inversée, la performance chute. La correction pertinente consiste à supprimer la variable non causale et à imposer une évaluation shifted.
